### 2.2 Prioritization Framework

Apply a multi-criteria prioritization approach.

```python
# YOUR CODE: Score initiatives on multiple criteria

# Score each initiative 1-5 on these criteria
scoring = {
    'Initiative': [],  # Match names from above
    'Business Value': [],
    'Strategic Fit': [],
    'Feasibility': [],
    'Time to Value': [],
    'Risk Level': [],  # Lower is better
}

df_scoring = pd.DataFrame(scoring)

# Calculate weighted score
weights = {
    'Business Value': 0.30,
    'Strategic Fit': 0.25,
    'Feasibility': 0.20,
    'Time to Value': 0.15,
    'Risk Level': 0.10
}

df_scoring['Weighted Score'] = (
    df_scoring['Business Value'] * weights['Business Value'] +
    df_scoring['Strategic Fit'] * weights['Strategic Fit'] +
    df_scoring['Feasibility'] * weights['Feasibility'] +
    df_scoring['Time to Value'] * weights['Time to Value'] +
    (6 - df_scoring['Risk Level']) * weights['Risk Level']  # Invert risk
)

df_scoring = df_scoring.sort_values('Weighted Score', ascending=False)
print("\nPrioritized Initiatives:")
print(df_scoring.to_string(index=False))
```

### 4.3 ROI Analysis

```python
# YOUR CODE: Calculate NPV, IRR, Payback Period

import numpy as np
from scipy.optimize import fsolve

# Cash flows (negative for costs, positive for benefits)
year_0 = 0  # Initial investment
year_1_net = 0  # Benefits - Costs
year_2_net = 0
year_3_net = 0

cash_flows = [year_0, year_1_net, year_2_net, year_3_net]
discount_rate = 0.10  # 10% WACC

# Calculate NPV
npv = sum([cf / (1 + discount_rate) ** i for i, cf in enumerate(cash_flows)])

# Calculate IRR
def npv_func(rate):
    return sum([cf / (1 + rate) ** i for i, cf in enumerate(cash_flows)])

irr = fsolve(npv_func, 0.1)[0] if sum(cash_flows) > 0 else 0

# Calculate Payback Period
cumulative = np.cumsum(cash_flows)
payback = next((i for i, v in enumerate(cumulative) if v > 0), None)

print("Financial Metrics:")
print(f"NPV @ {discount_rate*100:.0f}%: ${npv:,.0f}K")
print(f"IRR: {irr*100:.1f}%")
print(f"Payback Period: {payback if payback else '>3'} years")
print(f"ROI: {((sum(cash_flows[1:]) - abs(cash_flows[0])) / abs(cash_flows[0]) * 100):.1f}%")
```
